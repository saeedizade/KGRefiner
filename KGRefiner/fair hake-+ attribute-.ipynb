```python
path = '../../../LinkPrediction/Datasets/SACN/'
e2id = open(path + 'entity2id.txt').readlines()[1:]
e2id = [i[:-1].split('\t') for i in e2id]
e2id = [[str(int(j)),i] for i,j in e2id]
f = open('entities.dict','w')
for i,j in e2id:
  f.write(i+'\t'+j+'\n')
f.close()
```

```python
r2id = open(path + 'relation2id.txt').readlines()[1:]
r2id = [i[:-1].split('\t') for i in r2id]
r2id = [[str(int(j)),i] for i,j in r2id]
f = open('relations.dict','w')
for i,j in r2id:
  f.write(i+'\t'+j+'\n')
f.close()
```

```python
r2id = {i:j for i,j in r2id}
e2id = {i:j for i,j in e2id}
```

```python
train = open(path + 'train2id.txt').readlines()[1:]
train = [i[:-1].split(' ') for i in train]
train = [[e2id[i],r2id[k],e2id[j]] for i,j,k in train]
f = open('train.txt','w')
for i,j,k in train:
  f.write(i+'\t'+j+'\t'+k+'\n')
f.close()
```

```python
train = open(path + 'test2id.txt').readlines()[1:]
train = [i[:-1].split(' ') for i in train]
train = [[e2id[i],r2id[k],e2id[j]] for i,j,k in train]
f = open('test.txt','w')
for i,j,k in train:
  f.write(i+'\t'+j+'\t'+k+'\n')
f.close()
```
